```python
import random
from faker import Faker
import datetime
def generate_house_description(location):
    """
    Belirtilen sayıda kısa ev ilanı açıklaması üretir.
    """
    descriptions = []

    # Emlak ilanlarında sıkça geçen anahtar kelimeler ve ifadeler
    features = [
        "geniş salonlu", "aydınlık", "ferah", "deniz manzaralı", "şehir merkezine yakın",
        "ulaşım kolaylığı olan", "yeni yapım", "bakımlı", "eşyalı", "eşyasız",
        "güvenlikli site içinde", "havuzlu", "otoparklı", "balkonlu", "teraslı",
        "ankastre mutfaklı", "doğalgazlı", "klimalı", "ebeveyn banyolu", "giyinme odalı",
        "kombili", "jeneratörlü", "kapalı otoparklı", "çocuk parkına yakın", "okula yakın",
        "hastaneye yakın", "AVM'ye yakın", "sessiz ve sakin", "yeşillikler içinde",
        "deprem yönetmeliğine uygun", "ısı yalıtımlı", "ses yalıtımlı", "amerikan mutfak",
        "bahçe katı", "üst kat daire", "ara kat", "çatı dubleksi", "teras dubleksi"
    ]
    lifestyle_traits = [
        "temiz ve düzenli", "sakin bir yaşam seven", "sosyal ama özel alana saygılı",
        "hafta içi yoğun çalışan", "hafta sonu evde vakit geçirmeyi seven",
        "müzik dinlemeyi seven", "kitap okumayı seven", "yemek yapmaya meraklı",
        "sporla ilgili", "hayvansever", "doğa yürüyüşleri yapan", "sanatla ilgili",
        "mizah anlayışı yüksek", "uyumlu", "saygılı", "sorumluluk sahibi", "sigara kullanmayan",
        "alkol kullanımı olmayan", "misafir ağırlamayı seven", "erken kalkan", "gececi",
        "bitki seven", "minimalist", "çevreye duyarlı"
    ]

    # Evde beklentiler ve kurallar
    expectations = [
        "ortak alanların temizliğine önem veren", "faturaları düzenli ödeyen",
        "sessizliğe dikkat eden", "birbirinin eşyasına saygı duyan",
        "kişisel alanı önemseyen", "güleryüzlü bir ortam arayan",
        "samimi bir arkadaşlık kurmak isteyen", "sorunları konuşarak çözebilen",
        "evcil hayvanlara karşı hassas", "sigara içilmeyen bir ev arayan",
        "düzenli ev toplantılarına açık"
    ]
    # Başlangıç ve bitiş ifadeleri
    starts = [
        " ev.", " daire.",
    ]
    endings = [
        " birini arıyorum.", " bir ev arkadaşı arıyorum.", " kişileri bekliyorum. iyi anlaşacağımızı umuyorum.",
        " kişiler benimle iletişime geçebilir."
    ]
   
    # Şehrin 10 büyük ilçesini kullanarak konum ifadeleri oluşturma
    istanbul_districts = [
        "Ataşehir'de", "Başakşehir'de", "Bağcılar'da", "Bahçelievler'de", "Avcılar'da",
        "Beylikdüzü'nde", "Kadıköy'de", "Kartal'da", "Pendik'te", "Üsküdar'da"
    ]

    num_features = random.randint(1, 3)
    selected_features = random.sample(features, num_features)
    features_str = ", ".join(selected_features) + ("," if selected_features else "")
    
    num_exp = random.randint(1, 3)
    selected_exp = random.sample(expectations, num_exp)
    exp_str = ", ".join(selected_exp) + ("," if selected_exp else "")
    

    num_life = random.randint(1, 3)
    selected_life = random.sample(lifestyle_traits, num_life)
    life_str = ", ".join(selected_life) + ("," if selected_life else "")
    

    location_phrase = istanbul_districts[location]
    start_phrase = random.choice(starts)
    end_phrase = random.choice(endings)

    description = f"{location_phrase} {features_str}{start_phrase} { (exp_str+life_str+end_phrase if random.random() < 0.7 else '')}"
    
    # SQL'e uygun hale getirmek için tek tırnakları çift tırnağa çevir
    description = description.replace("'", "''")

    return description
```

```python


# Faker kütüphanesini Türkçe yerelleştirmesiyle başlat
fake = Faker('tr_TR')

# Oluşturulacak toplam kullanıcı ve ev sayısı
TOTAL_USERS = 500
USERS_WITH_HOMES = 250

# SQL sorgularını tutacak listeler
home_queries = []
registration_queries = []
attribute_queries = []
preference_queries = []

# İstanbul'un en büyük 10 ilçesi (önceki konuşmalarımızdan)
istanbul_districts = [
    "Ataşehir", "Başakşehir", "Bağcılar", "Bahçelievler", "Avcılar",
    "Beylikdüzü", "Kadıköy", "Kartal", "Pendik", "Üsküdar"
]

# --- 1. Adım: Ev Bilgilerini Oluşturma (250 Adet) ---
print(f"{USERS_WITH_HOMES} adet ev verisi oluşturuluyor...")
for i in range(1, USERS_WITH_HOMES + 1):
    homeID = i
    rent = round(random.uniform(5000, 50000), 2)
    
    # Konum için İstanbul ilçelerinden rastgele seçim
    loc =random.randint(0,len(istanbul_districts)-1)
    location = (istanbul_districts[loc]).replace("'", "''") 
    
    number_of_rooms = random.randint(1, 5)
    flat_floor = random.randint(0, 20)
    
    description_text = generate_house_description(loc).replace("'", "''")
    
    query = (
        f"INSERT INTO dbo.Home ( Rent, Location, NumberOfRooms, NumberOfSaloon, SurfaceArea, "
        f"Description, FlatFloor, BuildingFloor, BuildingFloorCount, Heating, BuildingAge) VALUES "
        f"({rent}, '{location}', {number_of_rooms}, {random.randint(0, 2)}, "
        f"{round(random.uniform(50, 250), 2)}, '{description_text}', "
        f"{flat_floor}, {random.randint(flat_floor, 25)}, {random.randint(1, 2)}, "
        f"'{random.choice(['Doğalgaz', 'Merkezi', 'Klima'])}', {random.randint(0, 30)});"
    )
    home_queries.append(query)

# --- 2. Adım: Kullanıcı Bilgilerini Oluşturma (500 Adet) ---
print(f"{TOTAL_USERS} adet kullanıcı verisi oluşturuluyor...")
for i in range(TOTAL_USERS):
    username = f"kullanici_{i+1}"
    first_name = fake.first_name().replace("'", "''")
    last_name = fake.last_name().replace("'", "''")
    phone_number = fake.phone_number()
    birth_date = fake.date_of_birth(minimum_age=18, maximum_age=65).strftime('%Y-%m-%d')
    password = fake.password(length=16, special_chars=False).replace("'", "''")
    profile_path = 'NULL'
    
    # --- RegistrationInfo Sorgusu ---
    reg_query = (
        f"INSERT INTO dbo.RegistrationInfo (username, firstName, lastName, phoneNumber, birthDate, password, ProfilePicturePath) "
        f"VALUES ('{username}', '{first_name}', '{last_name}', '{phone_number}', '{birth_date}', '{password}',{profile_path});"
    )
    registration_queries.append(reg_query)
    
    # Kullanıcının evi olup olmadığını belirle
    home_id_for_user = i + 1 if i < USERS_WITH_HOMES else 'NULL'

    # --- UserAttribute Sorgusu (Burada değişiklik yok, önceki haline göre kalır) ---
    # gender: 'm' veya 'f'
    # sleepSchedule: 'Sabah Kuşu' veya 'Gece Kuşu'
    # isSmoker/isDrinker: 0 (kullanmaz) veya 1 (kullanır)
    attr_query = (
        f"INSERT INTO dbo.UserAttribute (username, gender, sleepSchedule, isSmoker, isDrinker, Hygiene, "
        f"noisiness, homeID, GuestFrequency, IsStudent) VALUES "
        f"('{username}', '{random.choice(['m', 'f'])}', '{random.choice(['Sabah Kuşu', 'Gece Kuşu'])}', "
        f"{random.choice([0, 1])}, {random.choice([0, 1])}, {random.randint(1, 10)}, {random.randint(1, 10)}, "
        f"{home_id_for_user}, {random.randint(1, 10)}, {random.choice([0, 1])});"
    )
    attribute_queries.append(attr_query)

    # --- UserPreference Sorgusu (Burada 'Önemsiz' seçenekleri geri getirildi) ---
    min_budget = round(random.uniform(4000, 10000), 2)
    max_budget = min_budget + round(random.uniform(1000, 5000), 2)
    
    # desiredHomeLocation için 'Önemsiz' seçeneği eklendi
    desired_location = random.choice(istanbul_districts + ['Önemsiz']).replace("'", "''")
    
    # UserPreference için NoSmoker/NoDrinker mantığı (0 = fark etmez/içebilir, 1 = içmesin)
    # Python kodunda random olarak 0 veya 1 atayalım
    is_smoker_pref_db = random.choice([0, 1]) # 0 for "does not matter", 1 for "do not smoke"
    is_drinker_pref_db = random.choice([0, 1]) # 0 for "does not matter", 1 for "do not drink"

    pref_query = (
        f"INSERT INTO dbo.UserPreference (username, gender, sleepSchedule, NoSmoker, NoDrinker, minBudget, maxBudget, "
        f"desiredHomeLocation, SignificanceOfHygiene, NoiseSensitivity, GuestPreference, StudentPref) VALUES "
        f"('{username}', '{random.choice(['m', 'f', 'i'])}', '{random.choice(['Sabah Kuşu', 'Gece Kuşu', 'Önemsiz'])}', "
        f"{is_smoker_pref_db}, {is_drinker_pref_db}, {min_budget}, {max_budget}, '{desired_location}', "
        f"{random.randint(1, 10)}, {random.randint(1, 10)}, {random.randint(1, 10)}, {random.choice([0, 1])}); "
    )
    preference_queries.append(pref_query)

# --- 3. Adım: Tüm Sorguları Birleştirip Dosyaya Yazma ---
output_filename = 'olusturulan_veriler.sql'
with open(output_filename, 'w', encoding='utf-8') as f:
    f.write("-- Otomatik Oluşturulmuş SQL Verileri\n\n")
    f.write("-- Ev Bilgileri (" + str(len(home_queries)) + " adet)\n")
    f.write("\n".join(home_queries))
    f.write("\n\n-- Kayıt Bilgileri (" + str(len(registration_queries)) + " adet)\n")
    f.write("\n".join(registration_queries))
    f.write("\n\n-- Kullanıcı Özellikleri (" + str(len(attribute_queries)) + " adet)\n")
    f.write("\n".join(attribute_queries))
    f.write("\n\n-- Kullanıcı Tercihleri (" + str(len(preference_queries)) + " adet)\n")
    f.write("\n".join(preference_queries))

print(f"\nİşlem tamamlandı! '{output_filename}' dosyası başarıyla oluşturuldu.")
```
